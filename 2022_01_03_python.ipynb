## List, Dictionary, for loop

```python
# 1 ~ 100 까지의 숫자 중 홀수이면서 3의 배수를 리스트로 저장하시오(for loop사용)

a = []
for i in range(1, 101) :
    if((i%3 == 0) and (i%2 == 1)) : 
        a.append(i)
print(a)
```

```python
# 숫자를 반복적으로 입력 받아 계속 더해 갑니다. 더한 숫자가 30을 넘으면 반복문을 멈추고 총 합계를 출력합니다.

total = 0
while True :
    n = int(input())
    total += n
    if total > 30 :
        print(total)
        break
```

## Numpy

```python
# 리스트를 이용하여 2차원 배열을 생성하고

a = [[1,2,3,4,5],[4,5,6,7,8]]
```

```python
# shape
# 전체, 열기준(axis=0), 행기준(axis=1)
# 합, 평균, 표준편차를 구하시오

# 합
print(np.sum(a))
print(np.sum(a, axis=0))
print(np.sum(a, axis=1))

# 평균
print(np.mean(a))
print(np.mean(a, axis=0))
print(np.mean(a, axis=1))

# 표준편차
print(np.std(a))
print(np.std(a, axis=0))
print(np.std(a, axis=1))
```

* np.where
    * np.where(조건, True일 때 값, False일 때 값)

```python
# np.where를 이용하여 값이 짝수이면 1, 아니면 0을 출력하시오.

list1 = [[1,2,3,4,5],[4,5,6,7,8]]
a = np.array(list1)
print(a)
print(np.where(a%2 == 0, 1, 0))
```

* np.argmax : 가장 큰 값의 인덱스를 반환
* np.argmin : 가장 작은 값의 인덱스를 반환

```python
list1 = [[11,23,38], [25,31,19]]
a = np.array(list1)
print(a)
print()

print(np.argmax(a))
print(np.argmax(a, axis=0))
print(np.argmax(a, axis=1))
print()

print(np.argmin(a))
print(np.argmin(a, axis=0))
print(np.argmin(a, axis=1))
```

## Pandas

```python
# 데이터프레임에 대해서, 불러와서 상위 10개 행을 조회하세요.
path = 'https://bit.ly/3FsgwkJ'
titanic = pd.read_csv(path)
print(titanic.head(10))
```

```python
# 데이터프레임의 Age만 조회하시오

print(titanic['Age'])
```

```python
# 데이터프레임에서 1~5행까지, Name, Age, Embarked를 조회

print(titanic.loc[1:5], ['Name', 'Age', 'Embarked'])
```

```python
# Age가 70세 이상인 탑승객을 조회하시오

print(titanic.loc[titanic['Age']>=70])
```

```python
# Age가 60세 이상이고, 성별이 여성인 탑승객을 조회하시오

print(titanic.loc[(titanic['Age']>=60) & (titanic['Sex']=='female')])
```

```python
# Age가 20~22 상인 탑승객을 조회하시오

print(titanic.loc[titanic['Age'].between(20,22)])
```

```python
# Pclass가 1이거나 3인 탑승객을 조회하시오

print(titanic.loc[titanic['Pclass'].isin([1,3])])
```

```python
# 승선지역(Embarked)별 평균 운임을 조회하시오

print(titanic.groupby(by='Embarked', as_index=False)['Fare'].mean())
```

```python
# df1을 구성하는 data1, df2를 구성하는 data2가 아래와 같을 때, 두 데이터 프레임을 생성하고 행을 기준으로 병합하시오.

df1 = pd.DataFrame({'col1':[1,2], 'col2':['a','b']})
df2 = pd.DataFrame({'col1':[3,4], 'col2':['c','d']})
print(pd.concat([df1, df2], axis=0))
```

```python
# 앞서 생성한 데이터 프레임 df1, df2를 열을 기준으로 병합하시오

print(pd.concat([df1, df2], axis=1))
```
