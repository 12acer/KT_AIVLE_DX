* 날짜 정보 처리

* 날짜 정보 추출 방법
    * 문자열 형식의 날짜에서 년, 월, 일은 spit() 메소드를 사용해 얻을 수 있다.

```python
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/seoul_pop_h01.csv'
pop01 = pd.read_csv(path)
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/seoul_pop_h02.csv'
pop02 = pd.read_csv(path)
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/seoul_pop_h03.csv'
pop03 = pd.read_csv(path)
print(pop01)
print(pop02)
print(pop03)
```

```python
print(pop01.shape)
print(pop02.shape)
print(pop03.shape)
```

```python
print(pop01['year'].min())
print(pop01['year'].max())
print(pop02['year'].min())
print(pop02['year'].max())
print(pop03['year'].min())
print(pop03['year'].max())
```

```python
print(pop01.info())
print(pop02.info())
print(pop03.info())
```

```python
pop = pd.merge(pop01, pop02, on='year', how='outer')
pop = pd.merge(pop, pop03, on='year', how='outer')
print(pop)
```

```python
print(pd.isnull(pop).sum())
```

```python
pop.dropna(axis=0, inplace=True)
print(pop)
```

```python
pop['k_total'] = pop['k_male'] + pop['k_female']
pop['f_total'] = pop['f_male'] + pop['f_female']
pop['male'] = pop['k_male'] + pop['f_male']
pop['female'] = pop['k_female'] + pop['f_female']
pop['total'] = pop['k_total'] + pop['f_total']
print(pop.info())
```

```python
# pop.columns = ['year', 'household', 'total', 'male', 'female', 'k_total',
#              'k_male', 'k_female', 'f_total', 'f_male', 'f_female', 'older_65']
pop=pop[['year', 'household', 'total', 'male', 'female',
     'k_total', 'k_male', 'k_female', 'f_total', 'f_male', 'f_female', 'older_65']]

print(pop)
```

```python
pop.reset_index(drop=True, inplace=True)
print(pop)
```

```python
plt.plot(pop['year'], pop['total'])
plt.show()
```

```python
plt.plot(pop['year'], pop['male'])
plt.plot(pop['year'], pop['female'])
plt.show()
```

```python
plt.plot(pop['year'], pop['k_male'])
plt.plot(pop['year'], pop['k_female'])
plt.show()
```

```python
plt.plot(pop['year'], pop['f_male'])
plt.plot(pop['year'], pop['f_female'])
plt.show()
```

```python
plt.plot(pop['year'], pop['household'])
plt.show()
```

```python
plt.plot(pop['year'], pop['older_65'])
plt.show()
```
